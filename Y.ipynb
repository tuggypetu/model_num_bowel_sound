```python
def get_y(df):
    result = np.zeros(200)
    for i in range(len(df)):
        start = round(df.loc[i][0] * 100)
        end = round(df.loc[i][1] * 100) - 1 if round(df.loc[i][1] * 100) <= 200 else 199
        # print(start, end)
        frames_arr = np.arange(start, end + 1)
        for j in frames_arr:
            result[j] = 1
    # print(result)
    # print(len(result))
    return result
```

```python
df_Y = pd.DataFrame()
count = 0
for csv_file in csv_files:
    count += 1
    print(count, end=', ')
    fl_name = csv_file.split('/')
    fl_name = fl_name[-1].split('.')[0]
    df = pd.read_csv(csv_file)
    col = pd.Series(get_y(df))
    df_Y = pd.concat((df_Y, col.rename(fl_name)), axis=1)
df_Y
```

```python
df_Y = df_Y.sort_index(axis=1)
df_Y
```
